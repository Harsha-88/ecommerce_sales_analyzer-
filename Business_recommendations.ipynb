## Executive Summary

## Key Insights

1. Electronics is the top category
   Electronics gives the highest estimated revenue and also has the highest demand, so    it is the strongest category in this dataset.

2. Affordable products are more popular
   Most demand comes from products in the lower price range, especially 0–500, which      shows that customers prefer affordable products.

3. Discounts help increase demand
   Products with discounts between 40% and 80% have the highest demand. I noticed that    the 60–80% discount range performs especially well.

4. Higher discount does not mean better rating
   Even though discounts increase demand, products with higher discounts do not always    have better ratings. This shows that customers still care more about product     quality.

5. A few product types generate most revenue 
   Subcategories like Mobiles & Accessories and Home Entertainment generate a large    part of the revenue, which means only a few product groups are dominating the business.

## Business Risks

1. **Some products have high demand but low ratings**  
   I found some products that are selling well but their ratings are below 3.5. This can be risky because these products may create customer complaints, bad reviews, or returns.

2. **Electronics is strong but not best in ratings**  
   Electronics has the highest revenue and demand, but it does not have the highest average rating. This means the business is earning a lot from this category, but customer satisfaction can still be improved.

3. **Very high discounts may not always be good**  
   Products with very high discounts (80%+) do not show the highest demand and also have slightly lower ratings. This may reduce trust or show that some products are not very strong.

4. **Revenue depends a lot on a few categories**  
   A large part of the revenue is coming from Electronics and a few major subcategories. This means the business is depending a lot on limited product areas.

## Growth Opportunities

1. **Improve quality in Electronics**  
   Since Electronics is the most important category, improving product quality here can make a big difference in both customer satisfaction and long-term sales.

2. **Promote highly rated smaller categories**  
   Categories like Office Products, Toys & Games, and Home Improvement have better ratings. These categories may have good growth potential if they are promoted more.

3. **Focus more on good products with strong demand**  
   Products that have both high demand and good ratings should be given more focus because they are already performing well and customers also trust them.

4. **Keep affordable products in focus**  
   Since lower-priced products get more demand, the business can add more affordable products or bundle them with other items.

5. **Use discounts smartly**  
   The best demand is seen in the 40–80% discount range, especially around 60–80%, so this range can be used as a better pricing strategy.

## Final Recommendations
From analysis, I think the business should continue focusing on Electronics because it is the top category in both revenue and demand.

Overall, the best strategy is to improve quality in the main categories, keep affordable products in focus, and use discounts in a balanced way to increase both sales and customer trust.
