```python
import pandas as pd
df=pd.read_csv("../data/cleaned_sales_data.csv")
```

```python
import matplotlib.pyplot as plt

plt.hist(df['rating'], bins=20)

plt.title("Rating Distribution")
plt.xlabel("Rating")
plt.ylabel("Frequency")

plt.show()
```

The majority of products have ratings above 3.5, showing that overall customer satisfaction is high across the platform.

```python
df[['product_name','rating']].sort_values(by='rating', ascending=False).head(10)
```

```python
df[['product_name','rating']].sort_values(by='rating', ascending=True).head(10)
```

Highly rated products are mostly simple items like cables and small accessories.  
Low rated products are generally electronic gadgets like heaters and smart devices.

```python
df[['rating','rating_count']].corr()

```

There is a very weak positive relationship between rating and rating count.  
This means highly rated products may get slightly more demand, but ratings alone do not strongly drive sales.

```python
df[['rating','discounted_price']].corr()
```

There is a very weak positive relationship between rating and discounted price.  
This suggests that higher priced products may have slightly better ratings, but price is not a strong factor in customer satisfaction.

```python
df[['rating','discount_percentage']].corr()
```

There is a very weak negative relationship between rating and discount percentage.  
This indicates that products with higher discounts tend to have slightly lower ratings, but the impact is not strong.

### Correlation Summary

The correlation analysis shows that rating has only weak relationships with demand, price, and discount.  
Highly rated products do not always have the highest demand, expensive products are not necessarily much better rated, and higher discounts slightly reduce ratings.  
Overall, customer satisfaction seems to depend more on product quality than on price or discount strategy.

```python
df.groupby('main_category')['rating'].mean().sort_values()
```

Categories like Office Products and Toys & Games show the highest customer satisfaction, while Car & Motorbike and Musical Instruments have comparatively lower ratings.  
This suggests that some smaller categories may need quality improvement even if they are not major revenue drivers.
