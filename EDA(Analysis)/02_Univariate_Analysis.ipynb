```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
#load data
df=pd.read_csv("../data/cleaned_sales_data.csv")
df['actual_price'].head()
```

```python
df['actual_price'].describe()
```

```python
plt.hist(df['actual_price'], bins=30)
plt.title("Actual Price Distribution")
plt.xlabel("Actual Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.hist(df['actual_price'], bins=30)
plt.xlim(0,20000)
plt.title("Actual Price Distribution(0-20000)")
plt.xlabel("Price ")
plt.ylabel("No of products")
plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.boxplot(df['actual_price'], vert=True)

plt.title("Distribution of Actual Price")
plt.ylabel("Price (₹)")

plt.show()
```

### Observation (Data Insight)
- The histogram shows that most product prices are concentrated in the lower range.
- A large number of products fall between ₹0 and ₹5000, while very few products have prices above ₹20000.
- The boxplot confirms this distribution and shows many high-price outliers above the main price range.

### Business Insight
- The marketplace is primarily dominated by low to mid-priced products.
- A small number of premium products exist but they represent a minority of listings.
- This indicates that the platform mainly targets budget-conscious consumers while still offering some high-end products.

```python
plt.figure(figsize=(8,5))
df['main_category'].value_counts().head(10).plot(kind='bar')
plt.title("Product Distribution by Category")
plt.xlabel("Category")
plt.ylabel("Number of Products")

plt.show()
```
