```python
import pandas as pd

df=pd.read_csv("../data/cleaned_sales_data.csv")
```

```python
import matplotlib.pyplot as plt
df['estimated_revenue'] = df['discounted_price'] * df['rating_count']
```

```python
risk_products = df[(df['rating'] < 3.5) & (df['rating_count'] > df['rating_count'].median())]

risk_products[['product_name', 'main_category', 'rating', 'rating_count']].sort_values(by='rating').head(10)
```

Some products have high demand but low ratings, which shows a potential business risk.  
Many of these products belong to Computers & Accessories, suggesting possible quality or usability issues in this category.

```python
category_summary = df.groupby('main_category').agg({
    'estimated_revenue': 'sum',
    'rating': 'mean',
    'rating_count': 'sum'
}).sort_values(by='estimated_revenue', ascending=False)

```

Electronics generates the highest revenue and demand, but its average rating is lower than some smaller categories.  
This indicates a strong opportunity to improve product quality in the most important business segment.

```python
high_rating_low_revenue = category_summary[(category_summary['rating'] > category_summary['rating'].median())]

high_rating_low_revenue.sort_values(by='estimated_revenue', ascending=True)
```

Some smaller categories have higher average ratings but lower revenue.  
These categories may represent growth opportunities, as customers already show strong satisfaction and trust.

```python
over_discount = df[df['discount_percentage'] > 80]

over_discount[['product_name', 'main_category', 'discount_percentage', 'rating', 'rating_count']].head(10)
```

```python
over_discount.shape[0]
```

Products with extremely high discounts do not show the highest demand and tend to have slightly lower ratings.  
This suggests that excessive discounting may reduce customer trust or reflect weaker product quality.

```python
opportunity_products = df[(df['rating'] >= 4.3) & (df['rating_count'] > df['rating_count'].median())]

opportunity_products[['product_name', 'main_category', 'rating', 'rating_count', 'estimated_revenue']] \
.sort_values(by='estimated_revenue', ascending=False) \
.head(10)
```

Products with both high ratings and high demand represent strong growth opportunities.  
These products can be prioritized for promotion, inventory planning, and category expansion.

### Final Business Risks and Opportunities

This analysis highlights that some high-demand products have lower ratings, which creates a risk for customer trust and long-term retention.  
The Electronics category is the strongest revenue driver, but its ratings are lower than some smaller categories, showing a clear opportunity for quality improvement in the most important segment.  
At the same time, smaller categories like Office Products and Toys & Games have strong customer satisfaction, which suggests growth potential if promoted further.  
Very high discounts (80%+) do not appear to improve demand significantly and may reduce customer confidence, indicating a pricing risk.  
Overall, the best strategy is to improve quality in high-revenue categories, avoid excessive discounting, and scale highly rated products that already show strong demand.
