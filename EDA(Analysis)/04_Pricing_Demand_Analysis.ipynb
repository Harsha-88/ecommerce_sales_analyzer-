```python
import pandas as pd
df=pd.read_csv("../data/cleaned_sales_data.csv")
```

```python
import matplotlib.pyplot as plt
plt.hist(df['discounted_price'], bins=30)
plt.xlim(0,20000)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

Most products are in the lower price range.  
This shows that affordable products dominate the dataset.

```python
bins = [0, 500, 1000, 2000, 5000, 20000]
labels = ['0-500', '500-1000', '1000-2000', '2000-5000', '5000+']
df['price_range'] = pd.cut(df['discounted_price'], bins=bins, labels=labels)
df.groupby('price_range')['rating_count'].sum().sort_values(ascending=False)
```

Most demand is in the 0–500 price range.  
This shows that customers prefer low-priced products.

```python
plt.hist(df['discount_percentage'], bins=30)
plt.title("Discount Distribution")
plt.xlabel("Discount %")
plt.ylabel("Frequency")
plt.show()
```

The majority of products fall in the 30% to 70% discount range.  
This indicates that companies use moderate discounts as a strategy to attract customers.

```python
bins = [0, 20, 40, 60, 80, 100]
labels = ['0-20', '20-40', '40-60', '60-80', '80-100']
df.groupby('discount_range', observed=True)['rating_count'].sum().sort_values(ascending=False)
df['discount_range'] = pd.cut(df['discount_percentage'], bins=bins, labels=labels)
```

```python
df.groupby('discount_range', observed=True)['rating_count'].sum().sort_values(ascending=False)
```

Demand is highest for products with discounts between 40% and 80%.  
This shows that customers are more attracted to products with higher discounts.

```python
df.groupby('discount_range', observed=True)['rating'].mean()
```

Products with lower discounts have slightly higher ratings.  
This shows that ratings are not dependent on discounts.

### Final Conclusion
Demand is highest for low-priced products and products with moderate to high discounts (40%–80%).  
However, higher discounts do not lead to better ratings.  
This shows that while discounts attract customers, product quality is the main factor which affects ratings.
