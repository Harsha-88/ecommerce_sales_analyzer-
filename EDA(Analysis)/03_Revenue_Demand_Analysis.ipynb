```python
import pandas as pd
df=pd.read_csv("../data/cleaned_sales_data.csv")
```

```python
#Money earned = Price × Number of items sold
# we dont have no of items sold so we use rating_count as approximation
df['estimated_revenue']=df['discounted_price']*df['rating_count']
```

```python
df.columns
```

```python
df[['product_name','main_category','estimated_revenue']] \
.sort_values(by='estimated_revenue', ascending=False) \
.head(10)
```

Most of the top revenue products belong to the Electronics category.  
This shows that electronics products contribute significantly to overall revenue.

```python
import matplotlib.pyplot as plt
df.groupby('main_category')['estimated_revenue'].sum().sort_values(ascending=False).plot(kind='bar')

plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Estimated Revenue")

plt.show()
```

Electronics category generates the highest revenue compared to other categories.  
This shows that electronics products dominate overall sales performance.

```python
df.groupby('sub_category')['estimated_revenue'].sum().sort_values(ascending=False).plot(kind='bar')

plt.title("Top Subcategories by Revenue")
plt.xlabel("Subcategory")
plt.ylabel("Estimated Revenue")

plt.show()
```

Mobile and accessories contribute the most revenue within Electronics.  
This shows high demand for mobile-related products compared to others.

```python
df[['product_name','rating_count']] \
.sort_values(by='rating_count', ascending=False) \
.head(10)
```

Products like HDMI cables and earphones have the highest demand.  
This indicates that affordable and daily-use products are purchased more.

```python
df.groupby('main_category')['rating_count'] \
.sum() \
.sort_values(ascending=False) \
.plot(kind='bar')

plt.title("Demand by Category")
plt.xlabel("Category")
plt.ylabel("Total Rating Count")

plt.show()
```

Electronics category has the highest demand among all categories.  
This shows that most customers prefer electronics products over other products.

```python
df.groupby('sub_category')['rating_count'] \
.sum() \
.sort_values(ascending=False) \
.head(10) \
.plot(kind='bar')

plt.title("Top Subcategories by Demand")
plt.xlabel("Subcategory")
plt.ylabel("Total Rating Count")

plt.show()
```

Electronics category has the highest revenue as well as demand.  
Products like mobiles, HDMI cables and accessories contribute the most.

```python
df[['discounted_price','rating_count','estimated_revenue']].corr()
```

### Correlation Analysis Insight

Correlation analysis helps us understand the relationship between different variables like price, demand, and revenue.
From the results, we observe that there is a positive relationship between rating count (demand) and estimated revenue. This means that products which have more demand tend to generate higher revenue.
We also see a moderate positive relationship between discounted price and revenue. This indicates that higher priced products can contribute more to revenue.
However, there is almost no relationship between price and demand. This means that higher price does not necessarily reduce demand, and lower price does not always increase demand.
Overall, revenue is affected by both price and demand, but demand plays a more important role in driving revenue.
