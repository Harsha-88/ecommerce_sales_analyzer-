```python
import pandas as pd
df=pd.read_csv("../data/cleaned_sales_data.csv")
```

```python
import matplotlib.pyplot as plt
df['estimated_revenue'] = df['discounted_price'] * df['rating_count']
```

```python
import matplotlib.pyplot as plt

# group and calculate revenue
category_revenue = df.groupby('main_category')['estimated_revenue'].sum()

# sort values
category_revenue = category_revenue.sort_values(ascending=False)

# plot
category_revenue.plot(kind='bar')

plt.title("Revenue by Category")
plt.xlabel("Category")
plt.ylabel("Estimated Revenue")

plt.xticks(rotation=45)
plt.show()
```

Electronics category generates the highest estimated revenue compared to other categories. so it is a revenue drivers for our platform

```python
df.groupby('main_category')['rating_count'] \
.sum() \
.sort_values(ascending=False) \
.plot(kind='bar')

plt.title("Demand by Category")
plt.xlabel("Category")
plt.ylabel("Total Rating Count")

plt.xticks(rotation=45)
plt.show()
```

Electronics category has the highest rating count, which indicates higher demand compared to other categories.

```python
df.groupby('main_category')['rating'].mean().sort_values(ascending=False)
```

Electronics category has high demand and revenue but does not have the highest ratings.  
Some smaller categories like Office Products and Toys have higher customer satisfaction.

```python
df.groupby('sub_category')['estimated_revenue'] \
.sum() \
.sort_values(ascending=False) \
.head(10)
```

Mobiles and accessories generate the highest revenue among all subcategories.  
This shows that most customers spend on mobile-related products.

```python
df.groupby('sub_category')['rating_count'] \
.sum() \
.sort_values(ascending=False) \
.head(10)
```

Top subcategories by demand are similar to those generating the highest revenue.  
This confirms that high-demand products also contribute the most to revenue.
